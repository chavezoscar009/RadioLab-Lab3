```python
def data_acquisition(filename):
    
    '''
    Acquires the necessary data from the files given
    --------------------------------------------------------
    
    Parameters
    -----------
    filename(string): The filenmae of the data we want to get the data
    
    Output
    -----------
    volts: the voltage data from the filename
    times: the time series data from the filename
    freq: returns the frequency array for when we do power spectrum
    
    '''
    data = np.load(filename)
    
    volts = data['volts']
    times = data['times']
    
    freq = np.fft.fftfreq(len(times), d = 1)
    
    return volts, times, freq
```

```python
def power_spec(data):
    '''
    Self-explanatory: 
    This function calculates the power spectrum of the data
    -------------------------------------------------------
    
    Parameters
    ----------
    
    data: the data that we are going to be taking the power spectrum for
    
    Outputs
    ----------
    Power Spectrum
    '''
    fft = np.fft.fft(data)
    return np.abs(fft)**2
```

```python
def plot_power(freq, power, obj, x = 'None', y = 'None'):
    if x != 'None' and y != 'None':
        
        fig = plt.figure(figsize = (12,8))
        ax = fig.add_subplot(111)

        ax.set_title('Power Spectrum of ' + obj, fontsize = 12, fontweight = 'bold')
        ax.set_xlabel(r'Frequency $\nu$ [Hz]', fontsize = 12, fontweight = 'bold')
        ax.set_ylabel(r'Power $[mV^2 \cdot s^2]$', fontsize = 12, fontweight = 'bold')
        ax.set_xlim(-x, x)
        ax.set_ylim(0, y)
        ax.plot(np.fft.fftshift(freq), np.fft.fftshift(power))
        
    else:
        
        fig = plt.figure(figsize = (12,8))
        ax = fig.add_subplot(111)

        ax.set_title('Power Spectrum of ' + obj, fontsize = 12, fontweight = 'bold')
        ax.set_xlabel(r'Frequency $\nu$ [Hz]', fontsize = 12, fontweight = 'bold')
        ax.set_ylabel(r'Power $[mV^2 \cdot s^2]$', fontsize = 12, fontweight = 'bold')
        #ax.set_xlim(-x, x)
        #ax.set_ylim(0, y)
        ax.plot(np.fft.fftshift(freq), np.fft.fftshift(power))
        
```

```python
def ha_sun(time):
    
    '''
    This function calculates the hour angle for the sun data
    --------------------------------------------------------
    
    Parameters
    ------------
    
    time: unix time, this is the time returned in the data acquisition function
    
    Outputs
    ------------
    ha_: an array contianing the hour angle for the sun in radians
    '''
    
    ha_ = []
    for i in time:
        
        jd = ugradio.timing.julian_date(i)
        lst = ugradio.timing.lst(jd)
        ra, dec = ugradio.coord.sunpos(jd)
        
        h_s0 = lst - np.radians(ra)
        
        ha_.append(h_s0)
    
    return np.array(ha_)
```

```python
def ha_object(time, ra):
    
    '''
    This function calculates the hour angle for our object data
    --------------------------------------------------------
    
    Parameters
    ------------
    
    time: unix time, this is the time returned in the data acquisition function
    ra: the right ascension of the object we observed
    
    Outputs
    ------------
    ha_: an array contianing the hour angle for the sun in radians
    '''
    
    ha_ = []
    
    for i in time:
        
        jd = ugradio.timing.julian_date(i)
        lst = ugradio.timing.lst(jd)
        
        h_s0 = lst - np.radians(ra)
        
        ha_.append(h_s0)
    return np.array(ha_)
```

```python
def sun_declination(t):
    
    '''
    This function calculates the declination for the sun data
    --------------------------------------------------------
    
    Parameters
    ------------
    
    t: unix time, this is the time returned in the data acquisition function
    
    Outputs
    ------------
    dec: an array contianing the declination for the sun in radians
    '''
    dec = []
    
    for i in t:
        
        jday = ugradio.timing.julian_date(i)
        r, d = ugradio.coord.sunpos(jd=jday)
        
        dec.append(np.radians(d))
        
    return np.array(dec)    
```

Estimating the fringe frequencies until we can figure out what $B_{ew}$ and $B_{ns}$. 

```python
def fringe_freq(L, dec, h_s):
    
    '''
    Function that calculates the fringe frequency in our signal
    -----------------------------------------------------------
    
    Parameters
    ----------
    
    L: The latitude where we are at
    dec: the declination of the object
    h_s: the hour angle of the object
    
    Outputs
    ----------
    
    ff: fringe frequency in terms of cycle per radian
    ff_hr: fringe frequency in terms of cycle per hour
    ff_min: fringe frequency in terms of cycle per min
    ff_sec: fringe frequency in terms of cycle per second
    '''
    
    B_ew = 20 #m
    B_ns = 0
    
    lam = .025
    
    Q_ew = B_ew/lam *np.cos(dec)
    Q_ns = B_ns/lam * np.sin(L) * np.cos(dec)
    
    term1 = Q_ew * np.cos(h_s)
    term2 = Q_ns * np.sin(h_s)
    
    ff = term1 - term2
    
    ff_hr = ff * (np.pi / 12)
    ff_min = ff_hr / 60
    
    ff_sec = ff_min/60
    
    return  ff_hr, ff_min, ff_sec
```

What we are trying to do is to find the total delay so that we can find the baseline length between the two telescopes. 

$\tau_{tot} = \tau_g(h_s) + \tau_c$

We need to do a least Square fitting on the fringe equaiton given by:

$F(h_s) = Acos(2\pi \nu \tau_g') + Bsin(2 \pi \nu \tau_g')$

-----------
Minimization/ Least Square Fitting

$Q_{ew} = \frac{B_{ew}}{\lambda}cos\delta $ 
,
$Q_{ns} = \frac{B_{ns}}{\lambda} sinL cos\delta $ 

Recall that:

$\tau_g'(h_s) = [\frac{B_{ew}}{c}cos\delta]cos(h_s) + [\frac{B_{ns}}{c}sinLcos\delta]cos(h_s) = Q_{ew}cos(h_s) + Q_{ns}cos(h_s)$

---

Here we will guess values of $Q_{ew}$ and $Q_{ns}$ and we will pass in values of hour angles, which we calculated before.

```python
def least_square(h_s, ydata):
    '''
    Function that calculates the least square value residuals
    
    Parameters:
    ----------------------
    h_s: Hour angle of the object
    ydata: the voltage data of the object
    
    Returns:
    ----------------------
    S: matrix holding the residuals
    q_ns: matrix holding the values of Q_ns 
    q_ew: matrix holding the values of Q_ew
    A: matrix holding the values of A
    errA: matrix holding the values of errors on A
    B: matrix holding the values of B
    errB: matrix holding the values of the errors on B
    '''
    
    #initial variable declarations n is how big the size of linspace would be
    n = 1000
    
    #low and high represent the guessed at values of Q_ns and Q_ew
    #low = 1
    #high = 1000
    
    #Lists that has the guesses of Q_ns and Q_ew
    Q_ns = np.linspace(1, 50, n)
    Q_ew = np.linspace(400, 650, n)
    
    #making the model function that will go into our curve_fit function
    def model(y, A, B):
       
        return A*np.cos(2*np.pi*y) + B*np.sin(2*np.pi*y)
    
    #empty matrix of my residuals
    S = np.zeros((n,n))
    
    #empty matrix of my guessed values
    A = []
    B = []
    q_ns = np.zeros((n,n))
    q_ew = np.zeros((n,n))
    errA = np.zeros((n,n))
    errB = np.zeros((n,n))
    #for loop going through each value and calculating the residuals
    
    for i in range(len(Q_ew)):
        for j in range(len(Q_ns)):
            
            newy = Q_ew[i] * np.sin(h_s) + Q_ns[j] * np.cos(h_s)
            
            params, covar = curve_fit(model, newy, ydata)
            #print(covar)
            #print(np.diag(covar))
            dy_sq = (ydata - model(newy, params[0], params[1]))**2
            
            s = np.sum(dy_sq)
            
            A.append(params[0])
            B.append(params[1])
            S[i][j] = s
            q_ns[i][j] = Q_ns[j]
            q_ew[i][j] = Q_ew[i]
            errA[i][j] = np.sqrt(np.diag(covar))[0]
            errB[i][j] = np.sqrt(np.diag(covar))[1]
            
    return S, q_ns, q_ew, errA, errB, A, B
```

```python
S, q_ns, q_ew, err_ns, err_ew, A, B = least_square(sun_ha[3000:3500], sun_data[3000:3500])
```

```python
def model(A, B, Q_ew, Q_ns, h_s):
    '''
    Function that is supposed to model the fringe amplitude
    This is the F(h_s) in handout
    -------------------------------
    
    Parameters
    ----------
    A: the value of A we got from our fit
    B: the value of B we got from our fit
    Q_ew: the value of Q_ew we got from our fit
    Q_ns: the value of Q_ns we got from our fit
    h_s: the hour angle of the object
    
    Outputs
    ---------
    The fringe amplitude
    '''
    
    #calculating the y that will go into our fringe equation
    y = Q_ew * np.sin(h_s) + Q_ns * np.cos(h_s) 
    
    return A*np.cos(2*np.pi*y) + B*np.sin(2*np.pi*y)
```

```python
def solve_Qns(value, dec, L):
    '''
    Function that solves for the North-South Baseline distance
    --------------------------------
    
    Parameters
    ----------
    
    value: This is the value of Q_ns we got from our minimization
    dec: the declinaiton of the object in this case the sun
    L: latitude where we are at
    
    Output
    ----------
    
    Bns: the calculated North-South Baseline length in meters
    '''
    
    #value of lambda
    lam = .028 #m
    
    #formula to find the length of the NS baseline
    Bns = (value * lam) / (np.sin(L) * np.cos(dec))
    
    return Bns
```

Working on the Modulating function using the sum used in the lab report

```python
def function(M, N):
    '''
    This function generates the modulating function laid out in the lab report
    
    Parameters
    ----------
    M: This represents the ffR term in the modulating funciton and be sure to give it a wide range ~10
    N: This is the N in the sum and will sum little n from -N to +N
    Outputs
    ----------
    ffR: the range of values for ffR used in the modulating function
    value: The values of the modulating function
    '''
    
    #making a range of values to try out for our ff*R using the M variable as our starting and ending point
    ffR = np.linspace(-M, M, 500)
    
    #making a range of values to try out for our N values, we need this for the sum
    sum_list = np.linspace(-N , N, 500)
    
    #value is a list that will hold the value we get after computing the sum for each M value tried
    value = []
    
    for j in ffR:
        
        #this holds the addition of the sum we calculate below and resets to zero everytime we change our M value
        addition = 0
        
        for i in sum_list:
            #caluclates the sum and this is the equation used from the Lab report on MF theory
            #here i replaced ffR with j and n with i
            addition += np.sqrt(1-(i/N)**2) * np.cos(2*np.pi*j*i/N)
        value.append(addition)
    
    return ffR, value

M = 4
N = 10

ffR, value = function(M,N)

#me looking for the zeros in the modulating function
plt.figure(figsize = (12,8))
plt.plot(ffR, value)
plt.xlim(-M,M)
plt.ylim(-60,30)
plt.xlabel(r'$ff \cdot R$')
plt.ylabel('Modulating function')
plt.title('Modulating Function Theory')
plt.axvline(-3.13)
plt.axvline(-3.62)
```

```python
#getting my hour angle from my best guess of the zero point
ha1 = sun_ha[1430]
#position of where i think zero is in the modulating function
x1 = -3.62

#getting my hour angle from my best guess of the zero point
ha2 = sun_ha[4400]
#position of where i think zero is in the modulating function
x2 = -3.13

def theta_r(ha, x, B_ew):
    '''
    Function that will calculate the angular radius of the object given an hour angle
    and an x position you determine from the bessel function
    
    Parameters
    ----------
    ha: The hour angle at the given node in your sun/moon data
    x: the number you found the zero to be at in the bessel function
    B_ew: Your calculated EW Baseline
    
    Outputs
    ----------
    theta: The angular radius of the object in degrees 
    '''
    #value of lambda in meters
    lam = .028
    
    term = B_ew/lam *np.cos(ha)
    
    theta = x / (2*np.pi * term)
    
    return np.degrees(theta)

y1 = theta_r(ha1, x1, B_ew)
y2 = theta_r(ha2, x2, B_ew)

print(y1)
print(y2)
```
